```python
import pandas as pd
import numpy as np

# Create a sample dataset
np.random.seed(42)
data = pd.DataFrame({
    'Year': np.arange(2015, 2025),
    'Sales': np.random.randint(200, 500, 10),
    'Profit': np.random.randint(50, 200, 10),
    'Category': np.random.choice(['Electronics', 'Furniture', 'Clothing'], 10),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], 10)
})

data.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(data['Year'], data['Sales'], marker='o', color='b', label='Sales')
plt.plot(data['Year'], data['Profit'], marker='s', color='g', label='Profit')
plt.title("Yearly Sales and Profit")
plt.xlabel("Year")
plt.ylabel("Amount")
plt.legend()
plt.grid(True)
plt.show()




plt.figure(figsize=(8,5))
plt.bar(data['Year'], data['Sales'], color='orange')
plt.title("Sales by Year")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()



region_counts = data['Region'].value_counts()
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue','lightgreen','orange','pink'])
plt.title("Region Distribution")
plt.show()
```

```python
import seaborn as sns

sns.scatterplot(x='Sales', y='Profit', hue='Category', data=data, s=100)
plt.title("Sales vs Profit by Category")
plt.show()



corr = data[['Sales', 'Profit']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()



sns.boxplot(x='Category', y='Profit', data=data, palette='Set2')
plt.title("Profit Distribution by Category")
plt.show()
```

```python
import plotly.express as px

fig = px.line(data, x='Year', y='Sales', title='Interactive Sales Trend', markers=True)
fig.show()



fig = px.bar(data, x='Category', y='Profit', color='Category', title='Profit by Category')
fig.show()



fig = px.scatter(data, x='Sales', y='Profit', color='Region', size='Profit', hover_data=['Year'])
fig.update_layout(title='Sales vs Profit by Region')
fig.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.io import output_notebook

output_notebook()

source = ColumnDataSource(data)

p = figure(title="Bokeh: Sales vs Profit", x_axis_label='Sales', y_axis_label='Profit', width=700, height=400)
p.circle('Sales', 'Profit', size=10, color='navy', alpha=0.6, source=source)
show(p)
```
