```python
class cal:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def add(self):
        print(self.a+self.b)
    def sub(self):
        print(self.a-self.b)
    def mul(self):
        print(self.a*self.b)
    def div(self):
        print(self.a//self.b)
ob=cal(3,4)
ob.add()
ob.sub()
ob.div()
ob.mul()
```

```python
#operator overloading is adding 2 objects just like normal variables. 

class distance:

    def __init__(self, x, y):

        self.feet = x

        self.inches = y

    def __add__(self,other):  # Overloading +

        print("am in addition with + operator overloading")

        final_feet = self.feet + other.feet

        final_inches = self.inches + other.inches   # your custom logic

        if final_inches >= 12:

            final_feet += final_inches // 12

            final_inches = final_inches % 12

        return distance(final_feet, final_inches)

    def __str__(self):  # Overloading str()

        print("am in string magic method")

        return "distance: " + str(self.feet) + " feet and " + str(self.inches) + " inches"

d1 = distance(9, 10)

d2 = distance(8, 9)

d3=d1+d2  #

print(d3)
 
```

```python
# Task 2: Inheritance Examples

# 1. Multiple Inheritance
class A:
    def method_A(self):
        print("Method A")

class B:
    def method_B(self):
        print("Method B")

class C(A, B):
    def method_C(self):
        print("Method C")

obj = C()
obj.method_A()
obj.method_B()
obj.method_C()

print("\n")

# 2. Multilevel Inheritance
class Parent:
    def parent_method(self):
        print("Parent method")

class Child(Parent):
    def child_method(self):
        print("Child method")

class GrandChild(Child):
    def grandchild_method(self):
        print("GrandChild method")

obj2 = GrandChild()
obj2.parent_method()
obj2.child_method()
obj2.grandchild_method()

print("\n")

# 3. Hierarchical Inheritance
class Parent1:
    def parent1_method(self):
        print("Parent1 method")

class Child1(Parent1):
    def child1_method(self):
        print("Child1 method")

class Child2(Parent1):
    def child2_method(self):
        print("Child2 method")

c1 = Child1()
c2 = Child2()
c1.parent1_method()
c1.child1_method()
c2.parent1_method()
c2.child2_method()

print("\n")

# 4. Hybrid Inheritance (Multiple + Multilevel)
class Base:
    def base_method(self):
        print("Base method")

class D(Base):
    def method_D(self):
        print("Method D")

class E(Base):
    def method_E(self):
        print("Method E")

class F(D, E):
    def method_F(self):
        print("Method F")

obj3 = F()
obj3.base_method()
obj3.method_D()
obj3.method_E()
obj3.method_F()
```

```python
# Task 3: Overriding & Overloading

# Method Overriding
class Parent:
    def greet(self):
        print("Hello from Parent")

class Child(Parent):
    def greet(self):
        print("Hello from Child") 

obj = Child()
obj.greet()

print("\n")


class Math:
    def add(self, a, b=None):
        if b is not None:
            return a + b
        else:
            return a + 10  # default addition

math_obj = Math()
print(math_obj.add(5, 7))  # 12
print(math_obj.add(5))     # 15 (default)
```
